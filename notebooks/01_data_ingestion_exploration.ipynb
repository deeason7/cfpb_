```python
import pandas as pd
```

```python
data_path = "../data/raw/consumer_complaints.csv"
df = pd.read_csv(data_path, low_memory = False)
```

```python
# View first few rows
print("ðŸ”¹ Sample rows:")
display(df.head())
```

```python
# Overview
print("ðŸ”¹ Dataset Info:")
df.info()
```

```python
# Missing values
print("ðŸ”¹ Missing Values (in %):")
missing = df.isnull().mean().sort_values(ascending=False) * 100
display(missing)
```

```python
# Basic stats for numeric columns
print("ðŸ”¹ Descriptive Stats (Numerics):")
display(df.describe())
```

```python
# Unique values per column
print("ðŸ”¹ Unique Value Counts:")
display(df.nunique().sort_values(ascending=False))
```

```python
# Save summary
with open("../outputs/profiling/data_overview_summary.md", "w") as f:
    f.write("### Data Overview\n")
    f.write(str(df.info(buf=None)))
    f.write("\n\nMissing Values:\n")
    f.write(str(missing))  
```
